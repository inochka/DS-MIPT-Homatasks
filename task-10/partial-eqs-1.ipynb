<h1> Уравнение переноса </h1>

```python
N = 100 # x
M = 2000 # t

y = np.zeros((M, N))

y0 = np.zeros(N)
y0[0] = 1

h = 1 / N
tau = 1 / M 
c = 1

y[0, :] = y0
y[:, 0] = 1
#y[:, N - 1] = 0

print(y)
```

```python
for i in range(M - 1):
    for j in range(1, N):
        #print(j)
        if j == 0:
            y[i + 1, j] = y[i, j] - (c * tau / h) * ( y[i, j + 1] - y[i, j] )
            #print(y[i, j + 1] - y[i, j])
        elif j == N - 1:
            y[i + 1, j] = y[i, j] - (c * tau / h) * ( y[i, j] - y[i, j - 1] )
        else:
            y[i + 1, j] = y[i, j] - (c * tau / (2 * h) ) * ( y[i, j + 1] - y[i, j - 1] )
```

```python
print(y[M-1,:])
```

```python
print(y[1,:])
```

```python
sns.heatmap(y)
```

<h1> Волновое уравнение </h1>

```python
N = 50 # x
M = 10000 # t

y = np.zeros((M, N))

y0 = np.zeros(N)
y0[0] = 1

h = 1 / N
tau = 1 / M 
c = 1.0

y[0, :] = 0
y[1, :] = h * np.sin( 2 * np.linspace(0, 1, N) ) + h * np.sin( 3 * np.pi * np.linspace(0, 1, N) )

y[:, 0] = 0
y[:, N - 1] = 0

print(y)
```

```python
for i in range(1, M-1):
    for j in range(1, N-1):
        #print(y[i, j])
        y[i + 1, j] = 2 * y[i, j] - y[i - 1, j] + (c * tau * tau /  h / h) * ( y[i, j + 1] + y[i, j - 1] - 2 * y [i, j] )
```
